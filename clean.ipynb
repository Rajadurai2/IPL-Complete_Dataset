```python
import pandas as pd
import re
```

```python
df1=pd.read_csv('match_data.csv')
df=df1['winner'].values
print(df)
# df['win_team']=None
# df['win_by']=None
for k,i in enumerate(df,0):
    try:
        pattern = r"(.+) won by (.+)"
        team_name_match = re.search(pattern,i)
        super_over=re.search(r' (.+) won the (.+)',i)
        if team_name_match:
            df.loc[k,'win_team']=team_name_match.group(1)
            df.loc[k,'win_by']=team_name_match.group(2)
        elif super_over:
            
            df.loc[k,'win_team']=super_over.group(1).split('(')[-1]
            df.loc[k,'win_by']=super_over.group(2)
            
            
        else:
            print(i)
            df.loc[k,'win_team']=None
            df.loc[k,'win_by']=None
    
    except Exception as e:
        pass
display(df1)
```
